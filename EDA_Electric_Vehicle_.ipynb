```python
numerical_columns = ['Postal_Code', 'Model_Year', 'Electric_Range', 'Base_MSRP', 'Legislative_District', 'DOL_Vehicle_ID', '2020_Census_Tract']


categorical_columns = ['VIN_(1-10)', 'County', 'City', 'State', 'Make', 'Model', 'Electric_Vehicle_Type', 'CAFV_Eligibility', 'Vehicle_Location', 'Electric_Utility']

discrete_df = df.select_dtypes(include=['object'])
numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("-"*10, col_name, "-"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
```

```python
discrete_univariate_analysis(discrete_df)
```

```python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("-"*10, col_name, "-"*10)
        print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

```python
numerical_univariate_analysis(numerical_df)
```

# Visual Univariate Analysis on Numerical Columns

Frequency Distribution

```python
sns.set(style="whitegrid")# Univariate Analysis: Distribution of Numerical Columns

# Plot histograms for numerical columns

for column in numerical_columns:
    plt.figure(figsize=(15, 10))

    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

Outlier Detection

```python
# Box plots for numerical columns

for column in numerical_columns:
    plt.figure(figsize=(15, 10))

    sns.boxplot(x=df[column])
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```

```python
def describe_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"\
Column: {column}")
    print(f"Number of outliers: {len(outliers)}")
    print(f"Percentage of outliers: {len(outliers) / len(df) * 100:.2f}%")
    print(f"Range of outliers: {outliers[column].min()} to {outliers[column].max()}")
    print(f"Range of non-outliers: {df[(df[column] >= lower_bound) & (df[column] <= upper_bound)][column].min()} to {df[(df[column] >= lower_bound) & (df[column] <= upper_bound)][column].max()}")

for column in numerical_columns:
    describe_outliers(df, column)
```

```python
import pandas as pd
import plotly.express as px

# Step 1: Group by 'Model_Year' and 'Make', and calculate the count of EVs for each combination
ev_make_by_year = df.groupby(['Model_Year', 'Make']).size().reset_index(name='EV Count')

# Step 2: Create a list of all unique car makes
unique_makes = df['Make'].unique()

# Step 3: Ensure that all makes appear in every year by generating missing combinations
all_years = pd.DataFrame({'Model_Year': sorted(df['Model_Year'].unique())})
all_combinations = all_years.assign(key=1).merge(pd.DataFrame({'Make': unique_makes, 'key':1}), on='key').drop('key', axis=1)

# Step 4: Merge the combinations with the EV counts, filling missing values with 0
ev_make_by_year_full = all_combinations.merge(ev_make_by_year, on=['Model_Year', 'Make'], how='left').fillna(0)

# Step 5: Convert 'EV Count' column to integer (since missing values were filled as 0)
ev_make_by_year_full['EV Count'] = ev_make_by_year_full['EV Count'].astype(int)

# Step 6: Create the animated racing bar plot with increased height for better visibility
fig = px.bar(
    ev_make_by_year_full,  # Data source
    x='EV Count',  # X-axis for EV count
    y='Make',  # Y-axis for car make
    color='Make',  # Bar color by car make
    animation_frame='Model_Year',  # Animation by 'Model_Year'
    orientation='h',  # Horizontal bar chart
    title='Electric Vehicle Makes Over the Years',  # Title of the plot
    labels={'EV Count':'Number of EVs', 'Make':'Car Make'},  # Axis labels
    range_x=[0, ev_make_by_year_full['EV Count'].max() * 1.1],  # Dynamically set x-axis range with some padding
    height=800  # Increased height for better visibility
)

# Step 7: Show the plot
fig.show()
```
